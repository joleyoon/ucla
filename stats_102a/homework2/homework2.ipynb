```python
np.random.seed(406033530)
number_of_students = 100

# Generate unique UIDs safely
uids = np.random.choice(np.arange(100_000_000, 1_000_000_000), size=number_of_students, replace=False)

# Create gradebook
gradebook = pd.DataFrame({
    "UID": uids,
    "Homework1": np.random.randint(0, 101, number_of_students),
    "Homework2": np.random.randint(0, 101, number_of_students),
    "Homework3": np.random.randint(0, 101, number_of_students),
    "Homework4": np.random.randint(0, 101, number_of_students),
    "Homework5": np.random.randint(0, 101, number_of_students),
    "Quiz1": np.random.randint(0, 101, number_of_students),
    "Quiz2": np.random.randint(0, 101, number_of_students),
    "Quiz3": np.random.randint(0, 101, number_of_students),
    "Quiz4": np.random.randint(0, 101, number_of_students),
    "Quiz5": np.random.randint(0, 101, number_of_students),
    "Quiz6": np.random.randint(0, 101, number_of_students),
    "Quiz7": np.random.randint(0, 101, number_of_students)
})

gradebook.head()
```

```python
# Question 1b - Introduce Missing Values
hw1_to_remove = np.random.choice(gradebook.index, round(number_of_students * 0.1), replace=False)
hw5_to_remove = np.random.choice(gradebook.index, round(number_of_students * 0.1), replace=False)
quiz3_to_remove = np.random.choice(gradebook.index, round(number_of_students * 0.1), replace=False)

gradebook.loc[hw1_to_remove, "Homework1"] = np.nan
gradebook.loc[hw5_to_remove, "Homework5"] = np.nan
gradebook.loc[quiz3_to_remove, "Quiz3"] = np.nan

gradebook.head()
```

```python

# Question 1d - Apply Messy Imputation
print(messy_impute(gradebook, center="mean", margin=1).head())
print(messy_impute(gradebook, center="median", margin=2).head())

# Question 1e - Pivot to Tidy Format
gradebook_tidy = gradebook.melt(id_vars="UID", var_name="Assignment", value_name="Score")
gradebook_tidy.head()
```

```python
# Question 1g - Apply Tidy Imputation
print(tidy_impute(gradebook_tidy, center="mean", margin=1).head())
print(tidy_impute(gradebook_tidy, center="median", margin=2).head())
```

```python
# Question 2a - Load Data
romans = pd.read_csv("romans 2.csv")
sales = pd.read_csv("3.-Badly-Structured-Sales-Data-3.csv")
hurricanes = pd.read_csv("Hurricanes.csv")
```

```python
# Question 2c - Sample First 10-20 Rows and Save
sales_sample = sales.head(20)
romans_sample = romans.head(20)
hurricanes_sample = hurricanes.head(20)

sales_sample.to_csv("sample_sales_data.csv", index=False)
romans_sample.to_csv("sample_romans_data.csv", index=False)
hurricanes_sample.to_csv("sample_hurricanes_data.csv", index=False)

sales_sample.head(), romans_sample.head(), hurricanes_sample.head()
```

```python
# Extract 'Type' (column name of first non-NA value) and 'Value' (first non-NA value)
def first_non_na_name(x):
    non_na = x[x.notna()]
    return non_na.index[0] if len(non_na) > 0 else None

def first_non_na_value(x):
    non_na = x[x.notna()]
    return non_na.values[0] if len(non_na) > 0 else None

sales_sample_clean = pd.DataFrame({
    "Order ID": sales_sample.iloc[:, 0],
    "Order Date": sales_sample.iloc[:, 1],
    "Type": sales_sample.iloc[:, 2:].apply(first_non_na_name, axis=1),
    "Value": sales_sample.iloc[:, 2:].apply(first_non_na_value, axis=1)
})


# Roman Emperors Data Cleaning
# Split 'Birth' column into 'Birthdate' and 'Birthplace'
romans_sample_clean = romans_sample.copy()
romans_sample_clean[["Birthdate", "Birthplace"]] = romans_sample_clean["Birth"].str.split(",", n=1, expand=True)


# Hurricane Data Cleaning
# Separate 'Duration' into start/end day, then extract month/day/year
hurricanes_sample_clean = hurricanes_sample.copy()

# Split Duration into Start and End
hurricanes_sample_clean[["Start Day", "End Day"]] = hurricanes_sample_clean["Duration"].str.split(" – ", expand=True)

# Split Start Day into Month and Day
hurricanes_sample_clean[["Month", "Start Day Num"]] = hurricanes_sample_clean["Start Day"].str.split(" ", n=1, expand=True)

# Split End Day into Day and Year
hurricanes_sample_clean[["End Day Num", "Year"]] = hurricanes_sample_clean["End Day"].str.split(", ", n=1, expand=True)


# Preview Cleaned Data
sales_sample_clean.head(), romans_sample_clean.head(), hurricanes_sample_clean.head()
```
